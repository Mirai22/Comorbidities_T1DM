differnt version to find strings which works faster than the masking

```python
def search(regex: str, df, case=False):
    """Search all the text columns of `df`, return rows with any matches."""
    textlikes = df.select_dtypes(include=[object, "string"])
    return df[
        textlikes.apply(
            lambda column: column.str.contains(regex, regex=True, case=case, na=False)
        ).any(axis=1)
    ]
```

```python
df_T1DM = search("Diabetes Mellitus Type 1", df)
df_T1DM.reset_index(inplace=True)
df_T1DM = df_T1DM.drop('index', axis = 1)
df_T1DM
```

```python
df_T1DM_3 = search("T1DM", df)
df_T1DM_3.reset_index(inplace=True)
df_T1DM_3 = df_T1DM_3.drop('index', axis = 1)
df_T1DM_3
```

```python
df_T1DM_5 = search("Type 1", df)
df_T1DM_5.reset_index(inplace=True)
df_T1DM_5 = df_T1DM_5.drop('index', axis = 1)
df_T1DM_5
```

```python
df_list1 = [df_T1DM, df_T1DM_3, df_T1DM_5]
df_all_T1DMs = pd.concat(df_list1, ignore_index=True)
df_all_T1DMs.sort_values("SUBJECT_ID", inplace=True) 
df_all_T1DMs.drop_duplicates(subset=['SUBJECT_ID'],inplace = True)
df_all_T1DMs.reset_index(inplace=True)
df_all_T1DMs = df_all_T1DMs.drop('index', axis = 1)
df_all_T1DMs
```
