```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import re
```

```python
d_icd = pd.read_csv("DIAGNOSES_ICD.csv")  # contains icd codes with corresponding subject ids
d_icd_all = pd.read_csv("DIAGNOSES_ICD.csv")   
icd_names = pd.read_csv("D_ICD_DIAGNOSES.csv")  # contains icd codes with their coresponding names
```

```python
df = pd.read_csv("Comorb.csv")  # contains the one hot encoded comorbidity groups and subject ids of T1DM patients
typ1 = pd.read_csv("typ1.csv")  # contains the ogender, age and subject ids of T1DM patients
```

```python
merged_df = pd.merge(df, typ1, on='SUBJECT_ID')  # combine both preprocessed data
print(merged_df)
```

```python
# Suche nach NaN-Werten in der Spalte "Gender" und zeige die dazugehörigen "SUBJECT_ID"-Werte an
nan_gender_rows = merged_df[merged_df['Gender'].isna()]
nan_gender_subject_ids = nan_gender_rows['SUBJECT_ID']

# Zeige die "SUBJECT_ID"-Werte an, bei denen "Gender" NaN ist
print("SUBJECT_ID-Werte mit NaN in der Spalte 'Gender':")
print(nan_gender_subject_ids)
```

```python
# Duplikate in der Spalte "SUBJECT_ID" entfernen
unique_subject_ids_df = merged_df.drop_duplicates(subset=['SUBJECT_ID'])

# Das DataFrame mit eindeutigen "SUBJECT_ID"-Werten anzeigen
print(unique_subject_ids_df)
```

```python
# Mapping der Werte
gender_mapping = {'M': 1, 'F': 0}

# "Gender"-Spalte umwandeln und NaN-Werte beibehalten
unique_subject_ids_df['Gender'] = unique_subject_ids_df['Gender'].replace(gender_mapping)

# Das DataFrame mit den umgewandelten Werten anzeigen
print(unique_subject_ids_df)
```

```python
# Spalten "Gender_S" und "Gender_M" entfernen
columns_to_remove = ["Gender_S", "Gender_M"]
unique_subject_ids_df = unique_subject_ids_df.drop(columns=columns_to_remove)

# Das DataFrame ohne die entfernten Spalten anzeigen
print(unique_subject_ids_df)
```

```python
# Suche nach NaN-Werten in der Spalte "Gender" und zähle die Anzahl
nan_gender_count = unique_subject_ids_df['Gender'].isna().sum()

# Die Anzahl der NaN-Werte in der Spalte "Gender" anzeigen
print("Anzahl der NaN-Werte in der Spalte 'Gender':", nan_gender_count)
```

```python
# Suche nach NaN-Werten in der Spalte "Gender" und zeige die dazugehörigen "SUBJECT_ID"-Werte an
nan_gender_rows = unique_subject_ids_df[unique_subject_ids_df['Gender'].isna()]
nan_gender_subject_ids = nan_gender_rows['SUBJECT_ID']

# Zeige die "SUBJECT_ID"-Werte an, bei denen "Gender" NaN ist
print("SUBJECT_ID-Werte mit NaN in der Spalte 'Gender':")
print(nan_gender_subject_ids)
```
